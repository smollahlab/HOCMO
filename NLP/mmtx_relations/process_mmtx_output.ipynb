```python
import json
```

```python
diseases = [
    'breast_cancer',
    'cardiovascular_disease',
    'neurological_disease',
    'blood_cancer',
    'lung_cancer'
]
```

```python
relations_by_disease = {}
for disease in diseases:
    relations = {}
    for i in range(1,6):
        with open(f'{disease}{i}.json', 'r') as f:
            mmtx_output = json.load(f)
            for entry in mmtx_output:
                if entry['matchedtext'] not in relations:
                    relations[entry['matchedtext']] = entry
    relations_by_disease[disease] = relations
```

```python
for d, r in relations_by_disease.items():
    print(d, len(r))
```
